# Import Packages and Tools

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pyspark.sql import SparkSession

import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

```python
datapath = "DE_daily_sales.csv"
```

```python
df = pd.read_csv(datapath)
```

```python
df.info()
```

```python
spark_session = SparkSession \
                .builder \
                .appName('Technical Test MSBU') \
                .master("local[*]") \
                .getOrCreate()
```

```python
df = spark_session \
    .read \
    .option("header", True) \
    .csv(datapath)
```

```python
df.show()
```

```python
df.summary().show()
```

```python
df.printSchema()
```

```python
df = df.withColumn(
    "Transaction_Date",
    F.to_date(F.col("Transaction_Date"))
)
```

# Step 1: Data Cleaning

```python
df_preprocess = df.select('*')
```

```python
df_preprocess = df_preprocess.filter("Price IS NOT NULL")
# df_preprocess = df_preprocess.dropna(subset=['Price'])
```

```python
df_preprocess.count()
```

Can use another approach of missing data handling such as data statistic inputation (mean, medial, multivariate inputation) if needed

# Step 2: Tambah Kolom Baru Total_Revenue

```python
df_preprocess = df_preprocess.selectExpr(
    "*",
    "Price * Quantity AS Total_Revenue"
)

# df_preprocess = df_preprocess.withColumn(
#     "Total_Revenue",
#     F.col("Price") * F.col("Quantity")
# )

df_preprocess.show()
```

# Step 3: Pembuatan Dataset daily_sales_cleaned Beserta Query

```python
df_preprocess.createOrReplaceTempView("daily_sales_cleaned")

spark_session.sql("""
    SELECT 
        Transaction_Date,
        SUM(Total_Revenue) as Total_Revenue_Per_Day
    FROM daily_sales_cleaned
    GROUP BY Transaction_Date
    ORDER BY Transaction_Date
""").show()

# df_preprocess.groupby("Transaction_Date").agg(
#     F.sum("Total_Revenue").alias('Total_Revenue_Per_Day')
# ).sort("Transaction_Date").show()
```

```python
df_preprocess.toPandas().to_csv('daily_sales_cleaned.csv')
```

# EDA

## Trendline Total Revenue

### Per Day

```python
revenue_per_day = spark_session.sql("""
    SELECT 
        Transaction_Date,
        SUM(Total_Revenue) as Total_Revenue_Per_Day
    FROM daily_sales_cleaned
    GROUP BY Transaction_Date
    ORDER BY Transaction_Date
""").toPandas()
```

```python
plt.figure(figsize=[16, 8])

sns.lineplot(data=revenue_per_day, x="Transaction_Date", y="Total_Revenue_Per_Day")

plt.show()
```

### Per Month

```python
revenue_per_month = spark_session.sql("""
    SELECT 
        DATE_FORMAT(Transaction_Date, 'yyyy-MM') as Transaction_Date,
        SUM(Total_Revenue) as Total_Revenue_Per_Month
    FROM daily_sales_cleaned
    GROUP BY DATE_FORMAT(Transaction_Date, 'yyyy-MM')
    ORDER BY Transaction_Date
""").toPandas()
```

```python
plt.figure(figsize=[16, 8])

sns.lineplot(data=revenue_per_month, x="Transaction_Date", y="Total_Revenue_Per_Month")

plt.show()
```

## Top 20 Most Profitable Product

```python
revenue_per_product = spark_session.sql("""
    SELECT 
        Product_ID,
        SUM(Total_Revenue) as Total_Revenue_Per_Product
    FROM daily_sales_cleaned
    GROUP BY Product_ID
    ORDER BY Total_Revenue_Per_Product DESC
    LIMIT 20
""").toPandas()
```

```python
plt.figure(figsize=[16, 8])

colors = ['#1E4B7A']
colors.extend(['#4682B4'] * 2)
colors.extend(['#DEDEDE'] * (len(revenue_per_product) - 3))

sns.barplot(
    data=revenue_per_product, 
    y="Product_ID", 
    x="Total_Revenue_Per_Product",
    palette=colors
)

plt.show()
```

## Most Bought Product Each Month

```python
df_preprocess.show()
```

```python
# Find top product for each month
top_products_monthly = spark_session.sql("""
    WITH RankedProducts AS (
        SELECT 
            Month,
            Product_ID,
            Total_Revenue,
            ROW_NUMBER() OVER (PARTITION BY Month ORDER BY Total_Revenue DESC) as rank
        FROM (
            SELECT 
                DATE_FORMAT(Transaction_Date, 'yyyy-MM') as Month,
                Product_ID,
                SUM(Total_Revenue) as Total_Revenue
            FROM daily_sales_cleaned
            GROUP BY DATE_FORMAT(Transaction_Date, 'yyyy-MM'), Product_ID
        )
    )
    SELECT Month, Product_ID, Total_Revenue
    FROM RankedProducts
    WHERE rank = 1
    ORDER BY Month
""").toPandas()
```

```python
plt.figure(figsize=[16, 8])

colors = ['#1E4B7A', '#4682B4', '#4682B4'] + ['#DEDEDE'] * (len(top_products_monthly) - 3)
sns.barplot(
    data=top_products_monthly,
    x='Month',
    y='Total_Revenue',
    palette=colors
)

plt.title('Top Revenue Product by Month', pad=20, fontsize=14)
plt.xlabel('Month', labelpad=10)
plt.ylabel('Revenue', labelpad=10)
plt.xticks(rotation=45)

for i, row in enumerate(top_products_monthly.itertuples()):
    plt.text(i, row.Total_Revenue, f'Product: {row.Product_ID}', 
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
